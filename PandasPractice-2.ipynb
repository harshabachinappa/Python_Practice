ANALYSING DATAFRAMES

```python
x = [['Apple', '15 Pro', 2024, 67000], ['Samsung', 'S25 Ultra', np.nan, 79999], ['Google', 'Pixel6', 2025, 67666]]
df2 = pd.DataFrame(x, columns= ['Brand', 'Model', 'Year_of_Manufacture', 'Price'])
df2
```

```python
y = df2[df2.Price>=65000]
print(y)
y[['Brand', 'Price']]
```

```python
df2._append({'Brand': 'OnePlus', 'Model': 'Nord CE3', 'Year_of_Manufacture': 2024, 'Price': 12000}, ignore_index = True)
```

```python
df2._append({'Brand': 'Nokia', 'Model': '1100', 'Year_of_Manufacture': np.nan, 'Price': 2500}, ignore_index = True)
```
